An efficient way to check if a program construct is well-formed is to use a stack to scan the brackets once from left to right as follows:

Start from an empty stack. Whenever we encounter an open bracket, push it into the stack. Whenever we encounter a close bracket, we check if it is of the same type with the item at the top of the stack. Once we have a match, we pop the topmost bracket from the stack.
