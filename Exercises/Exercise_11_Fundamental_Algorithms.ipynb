# Exercise 11.1

Write the function `linearSearch1`, which takes in 2 parameters:

1. a list of integers, `L`, 
2. a target integer, `t`. 

This function will return the index of all instances of `t` in `L`. If `t` does not exist in `L`, then the function will return `-1`. Your implementation should be based on the linear search algorithm.

````python
#YOUR_CODE_HERE
def linearSearch1(L, t):
    indices = []
    for i, v in enumerate(L):
        if v == t:
            indices.append(i)
    return indices if indices else -1

array = [39, 96, 51, 20, 42, 42, 74, 28, 66, 16, 10, 86, 6, 43, 67, 98, 32, 73, 99, 7, 80, 88, 57, 83, 1, 64, 33, 38, 38, 8, 68, 38, 42, 80, 71, 82, 25, 29, 2, 85, 2, 96, 34, 14, 9, 65, 50, 63, 99, 94, 5, 93, 84, 46, 64, 22, 59, 31, 74, 13, 93, 13, 98, 93]

print(linearSearch1(array,9))
print(linearSearch1(array,2))
````

# Exercise 11.2

Write the function `binarySearch1`, which takes in 2 parameters:

1. a list of integers, `L`, which is sorted in ascending order  and 
2. a target integer, `t`. 

This function will return the index of any instance of `t` in `L`. If `t` does not exist in `L`, then the function will return `-1`. Your implementation should be based on the binary search algorithm.

````python
#YOUR_CODE_HERE
def binarySearch1(L, t):
    left, right = 0, len(L)-1
    
    while left <= right:
        mid = (left+right)//2
        if L[mid] == t:
            return mid
        elif L[mid] < t:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1

array = [39, 96, 51, 20, 42, 42, 74, 28, 66, 16, 10, 86, 6, 43, 67, 98, 32, 73, 99, 7, 80, 88, 57, 83, 1, 64, 33, 38, 38, 8, 68, 38, 42, 80, 71, 82, 25, 29, 2, 85, 2, 96, 34, 14, 9, 65, 50, 63, 99, 94, 5, 93, 84, 46, 64, 22, 59, 31, 74, 13, 93, 13, 98, 93]
sorted_array = sorted(array)
print(binarySearch1(sorted_array,39))
print(binarySearch1(sorted_array,96))
````

Write the function `linearSearch2`, which takes in 2 parameters: 

1. a list of integers, `L`, and 
2. a target integer, `t`. 

This function will return the index of any instance in `L` that is **just greater** than `t`.

We define an element `t'` in `L` is “just greater” than `t` if `t'` is greater than `t`, but less than (or equals) to any other element in `L` that is greater than `t`. If such any element does not exist in `L`, then the function will return `-1`. Your implementation should be based on the linear search algorithm.

````python
#YOUR_CODE_HERE
def linearSearch2(L, t):
    curr_iter = float('inf')
    curr_index = None
    for i,v in enumerate(L):
        if v < curr_iter and v > t:
            curr_iter = v
            curr_index = i

    return curr_iter,curr_index if curr_index else -1

array = [39, 96, 51, 20, 42, 42, 74, 28, 66, 16, 10, 86, 6, 43, 67, 98, 32, 73, 99, 7, 80, 88, 57, 83, 1, 64, 33, 38, 38, 8, 68, 38, 42, 80, 71, 82, 25, 29, 2, 85, 2, 96, 34, 14, 9, 65, 50, 63, 99, 94, 5, 93, 84, 46, 64, 22, 59, 31, 74, 13, 93, 13, 98, 93]

print(linearSearch2(array,9))
print(linearSearch2(array,2))
````

>```python

Write a script that will perform the following.

1. Repeatedly ask the user if they wish to perform a search. If they reply no, the script terminates, else the script continues.

2.  Asks the user to input one of the following search categories:
	- Overall score
	- Test score 1
	- Test score 2

	- Test score 4

3. Asks the user to input a search parameter (i.e., the score that they wish to search for).

4. Prints ALL the names of students that satisfy the search criteria. Note that you must implement an augmented version of linear search to perform this functionality. 

5. The script should then continue to the main loop in 1.

Your implementation must include all necessary exception handling.


````python
#YOUR_CODE_HERE
import csv
with open('resources/DATA.TXT') as f:
    reader = csv.reader(f)
    data = [row for row in reader]

while user_input:=input('Do you want to perform a search?').lower() != 'no':
    if (search_cat:=input('Input search category: ').lower()) in ['overall score', 'test score 1', 'test score 2', 'test score 3', 'test score 4']:
        if (score := input('Input search parameter: ')).isnumeric() and int(score) >= 0:
            student_list = []
            for student_name, test_score_1, test_score_2, test_score_3, test_score_4  in data:
                if search_cat == 'test score 1':
                    if score == test_score_1:
                        student_list.append(student_name)
                elif search_cat == 'test score 2':
                    if score == test_score_2:
                        student_list.append(student_name)
                elif search_cat == 'test score 3':
                    if score == test_score_3:
                        student_list.append(student_name)
                elif search_cat == 'test score 4':
                    if score == test_score_4:
                        student_list.append(student_name)
                elif search_cat == 'overall score':
                    if score == str((int(test_score_1) + int(test_score_2) + int(test_score_3) + int(test_score_4))/4):
                        student_list.append(student_name)
            print(f'{student_list}' if student_list else 'None found')
        else:
            print('Invalid input')
    else:
        print('Invalid input')
````
